**Employee Attrition**

It is the natural reduction in a company's workforce over time, typically occurring when employees leave voluntarily (like through retirement or resignation) and are not replaced. It differs from layoffs, which are employer-driven decisions to reduce staff. High attrition can negatively impact a company's productivity and stability, and is often a cause for concern.

```python
import pandas as pd
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
pd.options.display.max_columns = 500
```

```python
df
```

```python
#Binary Features: Gender, Attrition, Over18, OverTime
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x == "Yes" else 0)
df['Gender'] = df['Gender'].apply(lambda x: 1 if x == "Male" else 0)
df['Over18'] = df['Over18'].apply(lambda x: 1 if x == "Y" else 0)
df['OverTime'] = df['OverTime'].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
df
```

**One-Hot Encoding**



BusinessTravel, Department, EducationField, JobRole, Marital Status

```python
pd.get_dummies(df['BusinessTravel'])
```

```python
df = df.join(pd.get_dummies(df['BusinessTravel'])).drop('BusinessTravel', axis = 1)
```

```python
# Similarly for others
df = df.join(pd.get_dummies(df['Department'], prefix= 'Department')).drop('Department', axis = 1)
df = df.join(pd.get_dummies(df['EducationField'], prefix= 'Education')).drop('EducationField', axis = 1)
df = df.join(pd.get_dummies(df['JobRole'], prefix= 'Role')).drop('JobRole', axis = 1)
df = df.join(pd.get_dummies(df['MaritalStatus'], prefix= 'Status')).drop('MaritalStatus', axis = 1)
```

```python
df = df.map(lambda x: 1 if x is True else 0 if x is False else x)
```

```python
df
```

```python
df = df.drop('EmployeeNumber', axis = 1)
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df.hist(figsize=(20, 15))
plt.tight_layout()
plt.show()
```

```python
#dropping more features
df = df.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis = 1)
```

```python
df
```

**Model Training**

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X, Y = df.drop('Attrition', axis = 1), df['Attrition']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

model = RandomForestClassifier(n_jobs = -1)

model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
features_dict = dict(sorted(
    zip(model.feature_names_in_, model.feature_importances_),
    key=lambda x: x[1],
    reverse=True
))
```

```python
cleaned = {k: float(v) for k, v in features_dict.items()}
cleaned
```

```python
plt.figure(figsize=(20, 8))
plt.bar(cleaned.keys(), cleaned.values())
plt.xticks(rotation=90)
plt.show()
```
